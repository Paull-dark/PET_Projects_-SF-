```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
from tqdm import tqdm
import time  
from joblib import Parallel, delayed
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}
years = range(1950, 2021)

```

```python
'''Функция, возвращающая список производителей авто'''

def get_brands():    
    url = 'https://auto.ru/catalog/cars/'

    response = requests.get(url)
    response.encoding = 'utf-8'
    page = BS(response.text, 'html.parser')
    names_block = page.find(
        'div', class_='search-form-v2-mmm__content search-accordion__content')
    columns_block = names_block.find_all(
        'div', class_='search-form-v2-list__text-item')
    brands = [block.a['href'].split('/')[-2] for block in columns_block]
    return brands
```

```python
urls = [] # Список всех страниц

for brand in get_brands():
    for year in years:
        urls.append('https://auto.ru/moskva/cars/{}/{}-year/all/'.format(brand, year))
```

```python
def get_total_pages(url):
    """
    This function returns
    the last possible page 
    number for each car offers
    """
    response = requests.get(url, headers=headers)
    response.encoding = 'utf-8'
    pages = BS(response.text, 'html.parser').find_all('a',class_="Link OfferThumb")
    try:
        last_page = int(pages[-1].text)
        return last_page
    except:
        return 1


def get_cars_urls(url):
#     pdb.set_trace()
    all_urls = []
    
    last_page = get_total_pages(url)
    for page_i in range(1, last_page + 1):
        all_urls.append(f'{url}?page={str(page_i)}&output_type=list')
        
    return all_urls
        
            
    
```

```python
car_pages_lists_urls = Parallel(n_jobs=50)(delayed(get_cars_urls)(page_link) for page_link in tqdm(urls[900:905]))
```

```python

cars_list = Parallel(n_jobs=50)(delayed(cars)(page_link) for page_link in tqdm(urls))
```
